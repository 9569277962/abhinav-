```python
#answer 3
year= int(input("enter the year"))
if year%4==0 and year%100!=0:
    print(year,"is a leap year")
elif year%400==0:
    print(year,"is a leap year")
else:
    print(year,"is not leap year")
```

```python
#answer 4
n=int(input("number"))
c=0
i=1
while(i<=n):
    if(n%i==0):
        c=c+1
    i=i+1
if (c==2):
    print("prime number")
else:
    print("no prime")
```
